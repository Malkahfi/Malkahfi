```python
#Nomor 2
def fungsi (n):
    if n%2==1:
        b=int(n/2)
        
        print(" "*n + "- "*b +'panjang'+" -"*b)
        print('\r')
        for i in range(0,b):
            print(" "*4 + ' *   '*n)
            print('\r')

            print(' '*4+' *  '+'  =  '*(n-2)+'  *')
            print('\r')
        
        print(" "*4 + ' *   '*n)
        
    else:
        print("salah input. harus ganjil")
        
fungsi(7)
```

```python
#Nomor 6
def fungsi (n):
    if n%2==1:
        b=int(n/2)
        j=b

        #baris 1
        print(' '*(n+1)+'*')
        print('\r')

        for i in range (0,b-1):
            print(' '*2*j+'*'+' '*(2*i+1)+'*'+' '*(2*i+1)+'*')
            print('\r')
            j=j-1

        print(' '*2+'* '*n)
        print('\r')

        for i in range (b-2,-1,-1):
            j=j+1
            print(' '*2*j+'*'+' '*(2*i+1)+'*'+' '*(2*i+1)+'*')
            print('\r')

        #baris terakhir
        print(' '*(n+1)+'*')
        print('\r')
        
    else:
        print("salah input. harus ganjil")
        
fungsi(9)
```
